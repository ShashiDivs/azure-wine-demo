```python
wine = load_wine()

#  Create Dataframe
df = pd.DataFrame(
    data=wine.data,
    columns = wine.feature_names
)

df['target'] = wine.target

print(f"Dataset Loaded {df.shape[0]} samples, {df.shape[1]} features")
print(f"Target classes: {wine.target_names}")
```
